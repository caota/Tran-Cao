```python
import pandas as pd
df_can = pd.read_excel('Canada.xlsx',
                      sheet_name = 'Canada by Citizenship',
                      skiprows=range(20),
                      skipfooter=2)
```

```python
df_can.head()
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'],axis=1,inplace = True)
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'},inplace=True)
```

```python
df_can.head()
```

```python
all(isinstance(column, str) for column in df_can.columns)
```

```python
df_can.columns = list(map(str, df_can.columns))
```

```python
all(isinstance(column, str) for column in df_can.columns)
```

```python
df_can.set_index('Country', inplace=True)
```

```python
df_can.head()
```

```python
df_can['Total'] = df_can.sum(axis=1)
```

```python
df_can.head()
```

```python
years = list(map(str, range(1980, 2014)))

years
```

```python
%matplotlib inline 
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)
```

```python
# lấy mục nhập top 5
df_top5 = df_can.head()

# chuyển vị dataframe
df_top5 = df_top5[years].transpose() 

df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int) # hãy đổi giá trị chỉ mục của df_top5 thành kiểu integer để vẽ biểu đồ
df_top5.plot(kind='area', 
             stacked=False,
             figsize=(20, 10), # nhập kích thước tuple (x, y)
             )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_bottom5 = df_can.tail()[years].transpose()
df_bottom5.index = df_bottom5.index.map(int)
df_bottom5.plot(kind='area', 
             alpha=0.45, # 0-1, giá trị mặc định a= 0.5
             stacked=False,
             figsize=(10, 10),
            )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```
